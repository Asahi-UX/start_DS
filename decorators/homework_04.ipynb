## Импорт библиотек, установка констант

```python
import requests
import time
import re
from bs4 import BeautifulSoup

from random import randint
```

```python
BOOK_PATH = 'https://www.gutenberg.org/files/2638/2638-0.txt'
```

## Задание 1

```python
def benchmark(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Время выполнения функции {func.__name__}: {end_time - start_time}")
        return result
    return wrapper
```

## Задание 2

```python
def logging(func):
  def wrapper(*args, **kwargs):
    print(f"Функция вызвана с параметрами: {args}, {kwargs}")
    return func(*args, **kwargs)
  return wrapper

```

## Задание 3

```python
def counter(func):
    def wrapper(*args, **kwargs):
        wrapper.count += 1
        print(f"Функция {func.__name__} была вызвана {wrapper.count} раз")
        return func(*args, **kwargs)
    wrapper.count = 0
    return wrapper
```

## Задание 4

```python
def memo(func):
    cache = {}

    def fmemo(*args):
        key = tuple(args)
        if key in cache:
            return cache[key]
        else:
            result = func(*args)
            cache[key] = result
            return result

    return fmemo
```

## Тестирование

```python
@counter
@logging
@benchmark
def word_count(word, url=BOOK_PATH):

    raw = requests.get(url).text
    soup = BeautifulSoup(raw, 'html.parser')
    text = soup.get_text()

    # заменяем в тексте все небуквенные символы на пробелы
    processed_book = re.sub('[\W]+' , ' ', text).lower()

    # считаем
    cnt = len(re.findall(word.lower(), processed_book))

    return f"Cлово {word} встречается {cnt} раз"

print(word_count('whole'))

```

```python
@benchmark
def fib(n):
    if n < 2:
        return n
    return fib(n-2) + fib(n-1)
```

```python
# измеряем время выполнения
print(fib(20))
```

```python
@memo
def fib(n):
    if n < 2:
        return n
    return fib(n-2) + fib(n-1)
```

```python
# измеряем время выполнения
print(fib(20))
```
